```python
from sklearn.datasets import make_classification
import numpy as np
X,y=make_classification(n_samples=100,n_features=2,n_informative=1,
                        n_redundant=0,n_classes=2,n_clusters_per_class=1,
                        random_state=41,hypercube=False,class_sep=30)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

# Using Step Function

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1

    for i in range(1000):
        idx = np.random.randint(0,100)
        y_hat=step(np.dot(X[idx],weights))
        weights=weights + lr*(y[idx] - y_hat)*X[idx]

    return weights[0],weights[1:]
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
print(intercept_)
print(coef_)
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3,label='Step Function')
plt.scatter(X[:, 0],X[:,1],c=y,cmap='winter', s=100)
plt.ylim(-3,3)
plt.legend()
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3,label='Step Function')
plt.plot(x_input1,y_input1,color='black',linewidth=3,label='Logistic Regression')
plt.scatter(X[:, 0],X[:,1],c=y,cmap='winter', s=100)
plt.ylim(-3,3)
plt.legend()
```

# Using Sigmoid Function

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python

def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    weights=np.ones(X.shape[1])
    lr=0.1

    for i in range(1000):
        idx=np.random.randint(0,100)
        y_hat=sigmoid(np.dot(X[idx],weights))
        weights=weights + lr *(y[idx]-y_hat)*(X[idx])
    return weights[0],weights[1:]
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
x_input2=np.linspace(-3,3,100)
y_input2=m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3,label='Step Function')
plt.plot(x_input1,y_input1,color='black',linewidth=3,label='Logistic Regression')
plt.plot(x_input2,y_input2,color='yellow',linewidth=3,label='Sigmoid Function')
plt.scatter(X[:, 0],X[:,1],c=y,cmap='winter', s=100)
plt.legend()
plt.ylim(-3,3)
```
